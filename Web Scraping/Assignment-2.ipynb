```python
#job_titles extraction using for loop
job_title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in job_title_tags[0:10]:
    job_title.append(i.text)

#job_locations extraction
job_location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in job_location_tags[0:10]:
    job_location.append(i.text)
    
#company_name extraction
company_name_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_name_tags[0:10]:
    company_name.append(i.text)
    
#experiance extraction
experiance_req_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experiance_req_tags[0:10]:
    experiance_req.append(i.text)
```

```python
print(len(job_title),len(job_location),len(company_name),len(experiance_req))
```

```python
df=pd.DataFrame({'Job Title':job_title,'Job Location':job_location,'Company Name':company_name,'Experiance Required':experiance_req})
df
```

## Question 4

```python
driver=webdriver.Chrome(r'chromedriver.exe')
driver.get("http://www.flipkart.com/")
#To close login pop up
login=driver.find_element(By.XPATH,"/html/body/div[2]/div/div/button")
login.click()
#Search and submit to get data
products=driver.find_element(By.CLASS_NAME,"_3704LK")
products.send_keys('sunglasses')
search=driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']").click()
```

```python
brand_name=[]
product_description=[]
price=[]
discount=[]
for page in range(0,3):    
    brand_tags=driver.find_elements(By.XPATH,"//div[@class='_2WkVRV']")
    for i in brand_tags:
        brand=i.text
        brand_name.append(brand)

    product_tags=driver.find_elements(By.XPATH,"//a[@class='IRpwTa']")
    for i in product_tags:
        product=i.text
        product_description.append(product)    

    price_tags=driver.find_elements(By.XPATH,"//div[@class='_30jeq3']")
    for i in price_tags:
        rate=i.text
        price.append(rate)
    discount_tags=driver.find_elements(By.XPATH,"//div[@class='_3Ay6Sb']")
    for i in discount_tags:
        dis=i.text
        discount.append(dis)
        
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]').click()
    
    time.sleep(6)
```

```python
print(len(brand_name),len(product_description),len(price),len(discount))
```

```python
df=pd.DataFrame({'Brand':brand_name[0:100],'Product Description':product_description[0:100],'Price':price[0:100],'Discount':discount[0:100]})
df
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.amazon.in/')
laptop=driver.find_element(By.XPATH,'(//input[@class="nav-input nav-progressive-attribute"])[1]')
laptop.send_keys('Laptop')
search=driver.find_element(By.XPATH,'(//input[@class="nav-input nav-progressive-attribute"])[2]')
search.click()
#filter
filter=driver.find_element(By.XPATH,'(//i[@class="a-icon a-icon-checkbox"])[59]')
filter.click()
lap_name=[]
lap_price=[]
Rating=[]
laptop=driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-base a-text-normal"]')
for i in laptop[0:10]:
    lap_name.append(i.text)
laptop=driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in laptop[0:10]:
    lap_price.append(i.text)
titles_rat=driver.find_elements(By.XPATH,"//div[@class='a-row a-size-small']/span")
for i in titles_rat:
    Rating.append(i.get_attribute("aria-label"))
rating=[]
for i in range(0,len(Rating))[0:20]:
    if i == 0 or  i/2 == i//2:
        rating.append(Rating[i][0:3])
df=pd.DataFrame({'Laptop Name':lap_name,'Laptop Price':lap_price,'Ratings':rating})
df
```

## Question 8

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get('https://www.azquotes.com/')
#filter by top quotes
filter=driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
filter.click()
title=[]
author=[]
quotes=[]
for page in range(0,10):
    quote=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote:
        title.append(i.text)
    writer=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in writer:
        author.append(i.text)
    slogon=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in slogon:
        quotes.append(i.text)
```

```python
len(title),len(author),len(quotes)
```

```python
df=pd.DataFrame({'Quote':title,'Author':author,'Type of Quotes':quotes})
df
```

```python
Name=[]
born_dead=[]
term=[]
remark=[]
names = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[2]/p')
for name in names:
    Name.append(name.text)
born = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[3]/p')
for dob in born:
    born_dead.append(dob.text)
terms = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[4]/p')
for i in terms:
    term.append(i.text)
remarks = driver.find_elements(By.XPATH,'//div[@class="table-box"]/table/tbody/tr/td[5]/p')
for i in remarks:
    remark.append(i.text)
df=pd.DataFrame({'Name':Name,'Born-Dead':born_dead,'Remarks':remark})
df
```
