1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#lets first connect the driver
driver1=webdriver.Chrome("chromedriver.exe")
driver1.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Rank=[]
Video_Name=[]
Artist=[]
Date=[]
Views=[]
# scraping Rank of the videos
try:
    for i in driver1.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text.split('.')[0])
except NoSuchElementException:
    Rank.append("-")
# Scraping Name of the videos
try:
    for i in driver1.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Video_Name.append(i.text.split('"')[1])
except NoSuchElementException:
    Video_Name.append("-")
# Scraping Artist of the videos
try:
    for i in driver1.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
# Scraping Upload_Date of the videos
try:
    for i in driver1.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
# Scraping Views of the videos
try:
    for i in driver1.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
```

```python
len(Rank),len(Video_Name),len(Artist),len(Date),len(Views)
```

```python
df=pd.DataFrame({'Rank':Rank,'Video_Name':Video_Name,'Artist':Artist,'Upload Date':Date,'Views(In Billions)':Views})
df
```

4. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
#getting the specified url
driver4=webdriver.Chrome("chromedriver.exe")
driver4.get("https://github.com/")
```

```python
driver4.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button').click()
driver4.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a').click()
```

```python
rep_title=[]
url=[]
rep_desc=[]
contributors=[]
lang_used=[]
#for Repository Title
try:
    for i in driver4.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]'):
        rep_title.append(i.text)
except NoSuchElementException:
    rep_title.append("-")
#fetching urls for each repository
repository = driver4.find_elements(By.XPATH,"//h1[@class='h3 lh-condensed']//a")
for i in repository:
    url.append(i.get_attribute("href"))
# scraping data from all repository page
for i in url:
    driver4.get(i)
    
    # scraping Repository Description data 
    try:
        desc = driver4.find_element(By.XPATH,"//p[@class='f4 my-3']")
        rep_desc.append(desc.text)
    except NoSuchElementException:
        rep_desc.append('-')
        
    # scraping contributors count
    try:
        count = driver4.find_element(By.XPATH,"/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[3]/div[2]/div/div[5]/div/h2/a/span")
        contributors.append(count.text)
    except NoSuchElementException:
        contributors.append('-')
    except NameError:
        contributors.append('-')
    
    #scraping language used
    try:
        language=driver4.find_element(By.XPATH,'//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]')
        lang_used.append(language.text.replace('\n','-'))
    except NoSuchElementException:
        lang_used.append('-')
    
```

```python
df=pd.DataFrame({'Title':rep_title,'Descrition':rep_desc,'Contributors Count':contributors,'Language Used':lang_used})
df
```

```python
driver5.find_element(By.XPATH,"/html/body/div[3]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a").click()
driver5.find_element(By.XPATH,"/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[1]/div[1]/div[2]/span/a").click()
```

```python
song = []
artist = []
total_last = []
total_peak = []
weeks = []
for i in driver5.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//ul//li//ul//li[1]/h3'):
    song.append(i.text)
    
for i in driver5.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//ul//li//ul//li[1]/span'):
    artist.append(i.text)
    
for i in driver5.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//ul//li//ul//li[4]/span'):
    total_last.append(i.text)

last = total_last[::2]
    
for i in driver5.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//ul//li//ul//li[5]/span'):
    total_peak.append(i.text)
    
peak = total_peak[::2]
    
for i in driver5.find_elements(By.XPATH,'//div[@class="o-chart-results-list-row-container"]//ul//li//ul//li[6]/span'):
    weeks.append(i.text)
```

7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
#getting the specified url
driver7=webdriver.Chrome("chromedriver.exe")
driver7.get("https://www.imdb.com/list/ls095964455/")
```

```python
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []
# scraped data of Names
for i in driver7.find_elements(By.XPATH,"//h3[@class='lister-item-header']/a"):
    Name.append(i.text)
    
    
# scraped data of Year span
for i in driver7.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']"):
    Year_span.append(i.text)
    
    
# scraped data of Genre
for i in driver7.find_elements(By.XPATH,"//span[@class='genre']"):
    Genre.append(i.text)
    
    
# scraped data of Run time
for i in driver7.find_elements(By.XPATH,"//span[@class='runtime']"):
    Run_time.append(i.text)
    
    
# scraped data of Ratings
for i in driver7.find_elements(By.XPATH,"//div[@class='ipl-rating-star small']//span[2]"):
    Ratings.append(i.text)
    
    
# scraped data of Votes
for i in driver7.find_elements(By.XPATH,"//div[@class='lister-item-content']//p[4]/span[2]"):
    Votes.append(i.text) 
```

```python
df=pd.DataFrame({'Name':Name,
                'Year Span':Year_span,
                'Genre':Genre,
                'Run Time':Run_time,
                'Ratings':Ratings,
                'Votes':Votes})
df
```
